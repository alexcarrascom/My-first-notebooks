## Savings problem and parametrization
We use a standard instant CRRA utility function for household's preferences:
$$u(c) = \frac{c^{1-\gamma}}{1-\gamma}$$
Households discount future utilities stream by $\beta\in(0,1)$.

Household's income  $x$ is idiosyncratic and its log follows an AR(1) process, i.e.,
$$ \ln x_{t} = \rho \ln x_{t-1} + \epsilon_t, \quad \epsilon_t \sim \mathcal{N}(0,\sigma^2) $$
Then, the law of motion for the level of assets is:
$$a_t = Ra_{t-1} + x_t - c_t $$

**Household's Bellman equation**:
$$ V(a,x) = \max_{a'\in [0,Ra+x]} u[Ra+x-a'] + \beta\mathbb{E}[V(a',x')|x] $$

The household's state vector $\mathcal{S}$ is given by its level of asset and its income level, $\mathcal{S}=(a,x)$. The state of level of assets is numerically represented by a log-spaced grid of size $N$. Likewise, we discretize the idiosyncratic income process using Tauchen's method with $S$ discrete states. Therefore, after discretization, the numerical version of the Bellman equation is
$$ V(a_i,x_j) = \max_{a'\in [0,Ra_i+x_j]} u[Ra_i+x_j-a'] + \beta\sum_{s=1}^S V(a',x_s)P(j,s) $$
where $P$ is the stochastic matrix corresponding to the discrete Markov Chain approximation of $x$.

```julia
# Parameters
γ  = 2;
β  = 0.96;
R  = 1.01;
ρ  = 0.8;
σ  = 0.5;

N  = 100;
S  = 20;

# Utility Function
u(c)  = c > 0 ? c^(1-γ)/(1-γ) : -Inf;

# Numerical grids
mc    = QuantEcon.tauchen(S, ρ, σ);
P     = mc.p;
xgrid = exp.(mc.state_values) .+ 0.1;
agrid = exp.(range(log(1), stop = log(10), length = N)) .- 1;
```

## Serial/Direct Iteration

```julia
# Bellman step given evf and x
function bellman(evf, x, beta, R)
    vf, pol = similar(evf), similar(evf);
    for i in eachindex(vf)
        a             = agrid[i];
        obj           = aprime -> -(u(R*a - aprime + x) + beta * evf(aprime));
        upper         = min(R*a+x, agrid[end]);
        sol           = optimize(obj, agrid[1], .99*upper);
        vf[i], pol[i] = -sol.minimum, sol.minimizer;
    end
    return vf, pol;
end

function expectation(bigvf, P)
   evf = similar(bigvf)
    for a_nx in 1:size(bigvf, 1)
        for x_nx in 1:size(P, 1)
            evf[a_nx, x_nx] = sum(bigvf[a_nx, :] .* P[x_nx, :]);
        end
    end
    return evf
end
```

```julia
vfall  = zeros(length(agrid), length(xgrid))
polall = zeros(size(vfall));
function test_serial(vfall, polall, beta, R)
   fill!(vfall, 0); # reset the value function each time we benchmark
   fill!(polall, 0);
   for i in 1:100
      vfe = expectation(vfall, P);
      for x_nx in eachindex(xgrid)
         vfold          = interpolate(agrid, vfe[:, x_nx], SteffenMonotonicInterpolation());
         vf, pol        = bellman(vfold, xgrid[x_nx], beta, R);
         vfall[:,x_nx]  = vf; 
         polall[:,x_nx] = pol;
      end
   end
   return vfall, polall;
end
@time vfall, polall = test_serial(vfall, polall,β, R);    
p1=plot(agrid,vfall[:,[1, 6, 11, 16]],label=:none,lw=3); title!("Value Function (serial/direct)");
p2=plot(agrid,polall[:,[1, 6, 11, 16]].-agrid,label=:none,lw=3); title!("Saving Function (serial/direct)");
plot(p1,p2,layout=(1,2))
```

## Multi-Threading
Here, we use multithreading. See [here](https://docs.julialang.org/en/v1/manual/multi-threading/) for how to enable this. There is no interaction between loop iterations, so this is thread-safe. We check the output at the end.

First, lets see the number of execution threads:

```julia
Threads.nthreads()
```

```julia
vfall2  = zeros(size(vfall));
polall2 = zeros(size(vfall));

function test_threads(vfall, polall, beta, R)
    fill!(vfall, 0) # reset the value function each time we benchmark
    fill!(polall, 0)
    for i in 1:100
        vfe = expectation(vfall, P)
        Threads.@threads for x_nx in eachindex(xgrid)
            vfold          = interpolate(agrid, vfe[:, x_nx], SteffenMonotonicInterpolation())
            vf, pol        = bellman(vfold, xgrid[x_nx], beta, R)
            vfall[:,x_nx]  = vf; 
            polall[:,x_nx] = pol;
        end
    end
    return vfall, polall
end
@time vfall2, polall2 = test_threads(vfall2, polall2, β, R);    
p1=plot(agrid,vfall2[:,[1, 6, 11, 16]],label=:none,lw=3); title!("Value Function (thread)");
p2=plot(agrid,polall2[:,[1, 6, 11, 16]].-agrid,label=:none,lw=3); title!("Policy Function (thread)");
plot(p1,p2,layout=(1,2))
```

```julia
all(vfall .== vfall2) #did multithreading give the same result as serial?
```

**Result**: Multi-threading is 3x faster than the serial iteration of the value function.
